```python
from sklearn.preprocessing import StandardScaler
```

```python
df["sex"]=df["sex"].apply(lambda x: 1 if x == "male" else 0 )
df["region"]=df["region"].apply(lambda x: 0 if x == "southeast" else 1 if x=="southwest"else 2 if x=="northwest"else 3)
df["smoker"]=df["smoker"].apply(lambda x: 1 if x == "yes" else 0 )

```

```python
df
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
print(X)
print(y)

from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,y,random_state=2,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(Xtrain,Ytrain)
```

```python
ypred_train=model.predict(Xtrain)
from sklearn.metrics import r2_score
print("Linear model on train data",r2_score(Ytrain,ypred_train))
from sklearn.ensemble import BaggingRegressor,RandomForestRegressor,VotingRegressor
```

```python
model1=BaggingRegressor()
model1.fit(Xtrain,Ytrain)
ypred1_train=model1.predict(Xtrain)
from sklearn.metrics import r2_score
# r2_score(Ytrain,ypred1_train)
print("Bagging Regessor on train data: ",r2_score(Ytrain,ypred1_train))
# predicting on test data
ypred1_test=model1.predict(Xtest)
# from sklearn.metrics import r2_score
print("BaggingRegressor on test data: ",r2_score(Ytest,ypred1_test))
```

```python
model2=RandomForestRegressor()
model2.fit(Xtrain,Ytrain)
ypred2_train=model2.predict(Xtrain)
from sklearn.metrics import r2_score
r2_score(Ytrain,ypred2_train)
print("RandomForest Regressor on train data: ",r2_score(Ytrain,ypred2_train))
# predicting on test data
ypred2_test=model2.predict(Xtest)
# from sklearn.metrics import r2_score
print("RandomForest Regressor on test data: ",r2_score(Ytest,ypred2_test))
```

```python
model3=VotingRegressor(estimators=[("LinearRegression",model),("RandomForestRegressor",model2),("BaggingRegressor",model1)])
model3.fit(Xtrain,Ytrain)
ypred3_train=model3.predict(Xtrain)
from sklearn.metrics import r2_score
print("VotingRegressor on train data: ",r2_score(Ytrain,ypred3_train))
# predicting on test data
ypred3_test=model3.predict(Xtest)
# from sklearn.metrics import r2_score
print("VotingRegressor on test data: ",r2_score(Ytest,ypred3_test))
```
